```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as pex
import plotly.figure_factory as ff
from datetime import datetime
import os
```

```python
def addStack(df, n):
    l = len(df)
    Stack = [n for x in range(l)]
    df['Day'] = Stack
    return df

def addDates(sf, date):
    l = len(sf)
    df = sf.copy()
    #print(l)
    for i in range(l):
        s = df.iloc[i]['Start']
        f = df.loc[i]['Finish']
        s = date + ' ' + s
        f = date + ' ' + f
        #print(s, f)
        df.iloc[i]['Start'] = s
        df.iloc[i]['Finish'] = f
    return df
```

```python
startDate = '2021-10-21'  # Write your start date here
DFs = []
Days = len(os.listdir(r'..\CASAS_DATA\HH108\processed_data'))
for day in range(1, Days + 1):
    path = r'..\CASAS_DATA\HH108\processed_data\Day' + str(day) + '.csv'
    df = pd.read_csv(path)
    df = addDates(df, startDate)
    df = addStack(df, day)
    DFs.append(df)
    
```

```python
Full = DFs[0].copy()
Full = Full.append(DFs[1:])
Full.reset_index(inplace = True)
```

```python
gantt = pex.timeline(Full, x_start='Start', x_end='Finish', y='Day', color='Activity', height=600)
gantt
```
