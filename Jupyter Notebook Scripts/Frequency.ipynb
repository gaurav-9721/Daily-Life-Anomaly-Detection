```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
eat = ['Eat', 'Eat_Breakfast', 'Eat_Lunch', 'Eat_Dinner']
sleep = ['Sleep']
med = ['Morning_Meds', 'Evening_Meds', 'Medicine']
A = {'Eat': eat,  'Sleep': sleep, 'Medicine': med}
```

```python
def frequency_chart(dataset, acti):
    dirc = r'D:\Daily Life Anomaly Detection\CASAS_DATA\\' + dataset + r'\filling'
    Day_end = len(os.listdir(dirc))
    Day_start = 1
    E = []
    D = []
    for day in range(Day_start, Day_end+1):
        read_path = r'D:\Daily Life Anomaly Detection\CASAS_DATA\\' + dataset + r'\filling\Day' + str(day) + r'.csv'

        day_df = pd.read_csv(read_path)
        eat_df = day_df[day_df['Activity'].isin(A[acti])]
        E.append(len(eat_df))
        D.append(day)
    
    plt.plot(D, E, marker = 's', mfc = 'r', mec = 'r')
    plt.title('Frequencey of ' + acti + ' in ' + dataset)
    plt.xlabel('Days')
    plt.ylabel('Frequency')
```

```python
frequency_chart('HH101', 'Eat')
```
