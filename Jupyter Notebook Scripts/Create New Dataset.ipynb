```python
from random import randint

class Time:
    def __init__(self, timeStamp):
        self.time = timeStamp
        #self.hours, self.min, self.sec = map(int, timeStamp.split(':'))

    @staticmethod
    def splitTime( timeStamp):

        h, m, s = map(int, timeStamp.split(':'))
        return (h, m, s)

    @staticmethod
    def toSeconds(sec):
        h, m, s = Time.splitTime(sec)
        secs = h*3600 + m*60 + s
        return secs

    @staticmethod
    def toTime(sec):
        h = sec // 3600
        sec -= (3600 * h)
        m = sec // 60
        sec -= (60 * m)
        s = sec

        if h < 10:
            h = '0' + str(h)
        if m < 10:
            m = '0' + str(m)
        if s < 10:
            s = '0' + str(s)
        timeStamp = str(h) + ':' + str(m) + ':' + str(s)
        return timeStamp

    @staticmethod
    def randomTime( t1, t2):
        s1 = Time.toSeconds(t1)
        s2 = Time.toSeconds(t2)
        t = randint(s1, s2)
        t = Time.toTime(t)
        return Time(t)

    def __add__(self, other):
        return Time(self.toTime(Time.toSeconds(self.time) + Time.toSeconds(other.time)))

    def __sub__(self, other):
        max = Time.max(self, other).time
        min = Time.min(self, other).time
        return Time.toTime(Time.toSeconds(max) - Time.toSeconds(min))

    def __str__(self):
        return self.time

    def getTime(self):
        return self.time

    def __lt__(self, other):
        return Time.toSeconds(self.time) < Time.toSeconds(other.time)

    @staticmethod
    def max(t1, t2):
        if t1 < t2:
            return t2
        else:
            return t1

    @staticmethod
    def min(t1, t2):
        if t1 < t2:
            return t1
        else:
            return t2

if __name__ == '__main__':
    t1 = Time('15:01:00')
    t2 = Time('14:00:00')
    
```

```python
import pandas as pd
from random import randint
```

```python
def create_activity(activity, startMin, startMax, durMin, durMax, locations):
    Activity = []
    start = Time.randomTime(startMin, startMax)
    dur = Time.randomTime(durMin, durMax)
    finish = start + dur
    
    loc = randint(0, len(locations)-1)
    location = locations[loc]
    Activity.extend([activity, start.time, finish.time,  location])
    return Activity


def create_random_dataset():
    Schedule = []
    
    # First Argument is Activity Name
    # Second and third arguments are min and max Start Time. Random time between these 2 values will be chosen.
    # Fourth and Fifth arguments are min and max Start Time. Random time between these 2 values will be chosen.
    # Sixth argument is a list of Locations.
    
    sleep1 = create_activity('Sleep', '00:01:00', '01:01:00', '07:00:00', '08:10:00', ['Bedroom'])
    sleep2 = create_activity('Sleep', '23:00:00', '23:51:00', '00:10:00', '00:50:00', ['Bedroom'])
    Eat1 = create_activity('Eat', '08:00:00', '09:05:00', '00:30:00','00:30:00', ['Dining', 'Living Room'])
    Eat2 = create_activity('Eat', '12:10:00', '13:10:00', '00:12:00','00:19:00', ['Dining', 'Living Room'])
    Eat3 = create_activity('Eat', '19:00:00', '20:30:00', '00:15:00','00:20:00', ['Dining', 'Living Room'])
    medicine1 = create_activity('Morning_Meds','9:30:00', '10:30:00', '00:02:00', '00:02:30', ['Bedroom'])
    medicine2 = create_activity('Evening_Meds','21:10:00', '22:10:00', '00:02:00', '00:02:30', ['Bedroom'])
    
    sleep2[2] = '23:59:59'
    Schedule.extend([sleep1, Eat1, medicine1, Eat2, Eat3, medicine2, sleep2])
    df = pd.DataFrame(Schedule, columns=['Activity', 'Start', 'Finish', 'Location'])
    return df
    
    
```

```python
Days = 90

for day in range(1, Days+1):
    df = create_random_dataset()
    savepath = r'.\CASAS_DATA\HH000\processed_data\Day' + str(day) + '.csv' #Replace HH000 with another name
    df.to_csv(savepath, index= False)
    print(savepath, 'written successfuly.')
```
